```python
# i coded this with help of gpt
# register_user takes any number of sessions (*args) and user details (**kwargs)
# it prints and stores the user in the registrations list
registrations = []
def register_user(*args, **kwargs):
    print("User Registered:")
    print(f"Name: {kwargs["name"]}\nEmail: {kwargs["email"]}\nCountry: {kwargs["country"]}")
    sessions = list(args)
    print(f"Sessions: {sessions}")
    
    user_info = {
        "name": kwargs["name"],
        "email": kwargs["email"],
        "country": kwargs["country"],
        "sessions": sessions
    }

    registrations.append(user_info)
  
# show_all_registrations displays all registered users and their info
def show_all_registrations():
    print(f"Total Registrations: {len(registrations)}")

    for i, user in enumerate(registrations, start=1):
        print(f"\nUser #{i}")
        print(f"Name: {user['name']}")
        print(f"Email: {user['email']}")
        print(f"Country: {user['country']}")
        print(f"Sessions: {user['sessions']}")

register_user("Python Basics", "Data Science Bootcamp", name="Lina", email="lina93@gmail.com", country="Germany")
print("-------------------")
register_user("Web Dev", name="Ali", email="ali.dev@protonmail.com", country="Iran")
print("-------------------")
show_all_registrations()
```

```python
# check if email is already registered
# save data to json file

import json

registrations = []
def register_user(*args, **kwargs):
    print("User Registered:")
    print(f"Name: {kwargs["name"]}\nEmail: {kwargs["email"]}\nCountry: {kwargs["country"]}")
    sessions = list(args)
    print(f"Sessions: {sessions}")
    
    user_info = {
        "name": kwargs["name"],
        "email": kwargs["email"],
        "country": kwargs["country"],
        "sessions": sessions
    }

    # check for duplicate email
    for user in registrations:
        if user["email"] == kwargs["email"]:
            print("Already Registered!!")
            return
        
    registrations.append(user_info)

    # save to JSON file after each registration
    with open("file.json", "w") as f:
        json.dump(registrations, f)
    
  
def show_all_registrations():
    print(f"Total Registrations: {len(registrations)}")

    for i, user in enumerate(registrations, start=1):
        print(f"\nUser #{i}")
        print(f"Name: {user['name']}")
        print(f"Email: {user['email']}")
        print(f"Country: {user['country']}")
        print(f"Sessions: {user['sessions']}")

register_user("Python Basics", "Data Science Bootcamp", name="Lina", email="lina93@gmail.com", country="Germany")
print("-------------------")
register_user("Web Dev", name="Ali", email="ali.dev@protonmail.com", country="Iran")
print("-------------------")
show_all_registrations()
```

```python
# this version supports loading previous registrations from file (persistance)
# check if 'file.json' exists and loads data from it before starting

import os
import json
if os.path.exists("file.json"):
    with open("file.json", "r") as f:
        registrations = json.load(f)
else:
    registrations = []
    
def register_user(*args, **kwargs):
    print("User Registered:")
    print(f"Name: {kwargs["name"]}\nEmail: {kwargs["email"]}\nCountry: {kwargs["country"]}")
    sessions = list(args)
    print(f"Sessions: {sessions}")
    
    user_info = {
        "name": kwargs["name"],
        "email": kwargs["email"],
        "country": kwargs["country"],
        "sessions": sessions
    }
    for user in registrations:
        if user["email"] == kwargs["email"]:
            print("Already Registered!!")
            return
    registrations.append(user_info)
    with open("file.json", "w") as f:
        json.dump(registrations, f)
    
  
def show_all_registrations():
    print(f"Total Registrations: {len(registrations)}")

    for i, user in enumerate(registrations, start=1):
        print(f"\nUser #{i}")
        print(f"Name: {user['name']}")
        print(f"Email: {user['email']}")
        print(f"Country: {user['country']}")
        print(f"Sessions: {user['sessions']}")

register_user("Python Basics", "Data Science Bootcamp", name="Lina", email="lina93@gmail.com", country="Germany")
print("-------------------")
register_user("Web Dev", name="Ali", email="ali.dev@protonmail.com", country="Iran")
print("-------------------")
show_all_registrations()
```

```python
# CLI interface: allows interactive user input for registration and viewing
# wraps all functionality in a clean user menu with choices (1, 2, 3)

while True:
    print("\n==== Event Registration System ====")
    print("1. Register new user")
    print("2. Show all registrations")
    print("3. Exit")

    choice = input("Choose an option (1/2/3):")

    if choice == "1":
        name = input("Enter name: ")
        email = input("Enter email: ")
        country = input("Enter country: ")

        session_input = input("Enter sessions (comma-separated): ")
        sessions = [s.strip() for s in session_input.split(",") if s.strip()]

        register_user(*sessions, name=name, email=email, country=country)
    
    elif choice == "2":
        show_all_registrations()
    
    elif choice == "3":
        print("Exiting...")
        break

    else:
        print("Invalid choice. Please try again.")
```
