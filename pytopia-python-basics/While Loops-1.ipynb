```python
# gonna deposit $100 every month
# we have an annual interest rate of 5%
# how long does it take to reach $2000?

# looks like it’s around 10 months by eyeballing, but we need to calculate with a while loop
# need a while loop that keeps going until balance hits 2000
# also inside while we gotta check if 12 months passed, then apply interest
# at the end of 12 months → apply interest → balance = balance * 1.05

# and also need to print balance at the end of each year, so use while loop to track that too

# maybe one while loop is enough, or maybe we need two while loops
# one to apply deposits, and one to apply interest every 12 months
# or maybe it’s better to use one while + an if inside checking every 12 months

# and in bonus we gotta apply interest monthly or quarterly instead of yearly
```

```python
# without yearly interest

initial_balance = 1000
monthly_deposit = 100
annual_interest_rate = 0.05 
savings_goal = 2000

month_counter = 0 # do i need a month counter?

while initial_balance != 2000:
    initial_balance += monthly_deposit
    month_counter += 1
    print(month_counter, initial_balance)


print(f"It takes '{month_counter}' month to hit the target!!")
```

```python
# An improved version of the previous code.
# basic loop without interest
# Goal: count how many months to reach $2000 if we only deposit $100 per month (no interest applied)

initial_balance = 1000
monthly_deposit = 100
annual_interest_rate = 0.05
savings_goal = 2000

month_counter = 0  # counts how many months passed

while initial_balance < savings_goal: # why? because if it doesn’t hit exactly 2000, the loop will get stuck.
    initial_balance += monthly_deposit
    month_counter += 1
    print(month_counter, initial_balance)

print(f"It takes {month_counter} months to hit the target!")
```

```python
# Version 2: balance after one year, applying annual interest at end of 12 months

initial_balance = 1000
monthly_deposit = 100
annual_interest_rate = 0.05
savings_goal = 2000

months_of_year = 0 # counter for 12 months

while months_of_year < 12: # GPT says it’s clearer and easier to debug, but mine works too!
    initial_balance += monthly_deposit
    months_of_year += 1

# apply annual interest after 12 months

year_interest = initial_balance * annual_interest_rate
initial_balance += year_interest

print(initial_balance)

```

```python
# trying monthly interest + counting months until reaching savings goal
# Revised version of the previous one.

initial_balance = 1000
monthly_deposit = 100
annual_interest_rate = 0.05 
savings_goal = 2000

month_counter = 0
monthly_intreset_rate = annual_interest_rate / 12

while initial_balance < savings_goal:
    initial_balance += monthly_deposit
    initial_balance *= (1 + monthly_intreset_rate)
    month_counter += 1
    print(month_counter, initial_balance)

print(f"It takes {month_counter} months to hit the target!!!")
```

```python
# Final version: one loop, interest_type flag controls when to apply interest
# interest_type options : "yearly", "quarterly", "monthly"

initial_balance = 1000
monthly_deposit = 100
annual_interest_rate = 0.05
savings_goal = 2000

interest_type = "monthly" # choose : "yearly", "quarterly", "monthly"

month_counter = 0
months_to_apply = 12 # default yearly

if interest_type == "monthly":
    months_to_apply = 1
elif interest_type == "quarterly":
    months_to_apply = 3
# else stays as 12 for yearly

monthly_interest_rate = annual_interest_rate / 12

while initial_balance < savings_goal:
    initial_balance += monthly_deposit
    month_counter += 1

    if month_counter % months_to_apply == 0:
        # apply interest depending on frequency
        interest = initial_balance * (monthly_interest_rate * months_to_apply)
        initial_balance += interest
        print(f"End of month {month_counter}: balance = {initial_balance:.2f}")

print(f"Goal reached in {month_counter} months! Final balance: {initial_balance:.2f}")
```
